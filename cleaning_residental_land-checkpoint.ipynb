```python
df = land.drop(columns=['link','possession','property_id'])
```

```python
df.rename(columns={'floorNumAllowed' : 'floors'},inplace=True)
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','',str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df['price'].value_counts()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1]== 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df['areaWithType'] = df['areaWithType'].str.split(')').str.get(0).str.replace('(','').str.replace('sq.m.','').str.replace(')','').str.strip()
```

```python
df['floors_num'] = df['floors'].str.extract(r'(\d+)').astype(float)
```

```python
df['floors_num'] = df.groupby('society')['floors_num'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)
```

```python
df['floors_num'].isna().sum()
```

```python
df['sector'] = df['address'].str.extract(r'(Sector \d+)')
```

```python
sector_fill = df.groupby('sector')['floors_num'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)

df['floors_num'] = df['floors_num'].fillna(sector_fill)
```
